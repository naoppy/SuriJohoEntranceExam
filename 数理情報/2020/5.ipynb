```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 100, 100)
f1 = lambda x: (x - 30)**2
g1 = lambda x: 0 if x >= 30 else (x - 30)**2
p1 = p1 = [30, 0]
```

```python
plt.plot(x, f1(x), label='f1(x) = (x-30)^2')
plt.plot(x, list(map(g1, x)), label='g1(x)')
plt.plot(p1[0], p1[1], 'ro', label='p1')
plt.legend()
```

```python
f2 = lambda x: (x - 70)**2 if x >= 30 else 2 * x**2 - 200 * x + 5800
plt.plot(x[:30], list(map(f2, x[:30])), label="(x-30)^2 + (x-70)^2")
plt.plot(x[30:], list(map(f2, x[30:])), label="(x-70)^2")
plt.legend()
```

```python
g2 = lambda x: (
    0
    if x >= 70
    else (x - 70)**2 if x >= 30 else 2 * x**2 - 200 * x + 5800
)
p2 = [70, 0]
plt.plot(x[:30], list(map(g2, x[:30])), label='(x-30)^2 + (x-70)^2')
plt.plot(x[30:70], list(map(g2, x[30:70])), label='(x-70)^2')
plt.plot(x[70:], list(map(g2, x[70:])), label='0')
plt.plot(p2[0], p2[1], 'ro', label='p2')
plt.legend()
```
